```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
weight = float(input("Enter your weight (kg): "))
height_cm = float(input("Enter your height (cm): "))
```

```python
# Convert cm to meters first
height_m = height_cm / 100
#Calculate BMI Formula
bmi = weight/ (height_m * 2)
# Round to 1 decimal place like 22.9
bmi = round(bmi, 1)
print(f"Your BMI is: {bmi}")
```

```python
# Default values so python never gets confused
category = ""
feedback = ""

if height_cm == 0 or weight == 0:
    category = "Invalid Input"
    feedback = "Error: Weight and Height cannot be zero!"
elif bmi <18.5:
    category = "Underweight"
    feedback = "You're underweight. consider eating more nutrient-dense foods and strength training."
elif 18.5 <= bmi <= 24.9:
    category = "Normal weight"
    feedback = "Great job! Maintain your current nutrient of a balanced diet and regular excercise."
elif 25.0 <= 29.9:
    category = "Overweight"
    feedback = "You are slightly overweight.Add more veggies, water, and 30min walks daily."
elif 30.0 <= bmi <= 34.9:
    category = "Obese"
    feedback = "Obese category. Speak to a doctor/dietician for a safe weight loss pain."
elif 35.0 <= bmi <= 39.9:
    category = "Severly obese"
    feedback = "Severely obese. Medical guidance is strongly advised for your health."
else: # bmi >= 40
    category = "Morbidly obese"
    feedback = "Morbidly obese. Please consult a healthcare professional immediately."

print(f"Category: {category}")
print(f"Feedback: {feedback}")
    
```

```python
print("\n" + "="*50)
print(f"          BMI REPORT FOR {name.upper()}")
print("="*50)
print(f"Age: {age} years")
print(f"Height: {height_cm} cm | Weight: {weight} kg")
print(f"BMI score: {bmi}")
print(f"Category: {category}")
print(f"Advice: {feedback}")
print("="*50)
print("\nBMI Categories Reference:")
print("Underweight      : < 18.5")
print("Normal weight    : 18.5 - 24.9")
print("Overweight       : 25.0 - 29.9")
print("Obese            : 30.0 - 34.9")
print("Severly obese    : 35.0 - 39.9")
print("Morbidly obese   : >= 40")
print("="*50)
```

```python
print("\nLet's calculate BMI for 3 family members!\n")

for person in range(1, 4): # This means: do it 3 times
    print(f"--- Person {person} ---")

    name = input("Enter name: ")
    age = int(input("Enter age: "))
    weight = float(input("Enter weight in kg: "))
    height_cm = float(input("Enter height in cm: "))

    height_m = height_cm / 100
    bmi = round(weight / (height_m ** 2), 1)
    

    # Same category + feedback code from cell 4
    category = ""
    feedback = ""

    if height_cm == 0 or weight == 0:
        category = "Invalid Input"
        feedback = "Error: Weight and Height cannot be zero!"
    elif bmi <18.5:
        category = "Underweight"
        feedback = "You're underweight. consider eating more nutrient-dense foods and strength training."
    elif 18.5 <= bmi <= 24.9:
        category = "Normal weight"
        feedback = "Great job! Maintain your current nutrient of a balanced diet and regular excercise."
    elif 25.0 <= 29.9:
        category = "Overweight"
        feedback = "You are slightly overweight.Add more veggies, water, and 30min walks daily."
    elif 30.0 <= bmi <= 34.9:
        category = "Obese"
        feedback = "Obese category. Speak to a doctor/dietician for a safe weight loss pain."
    elif 35.0 <= bmi <= 39.9:
        category = "Severly obese"
        feedback = "Severely obese. Medical guidance is strongly advised for your health."
    else: # bmi >= 40
        category = "Morbidly obese"
        feedback = "Morbidly obese. Please consult a healthcare professional immediately."

    

    print("="*40)
    print(f"BMI Report for {name.upper()}")
    print(f"Age: {age} | BMI:{bmi} | Category: {category}")
    print(f"Advice: {feedback}")
    print("="*40 + "\n")
    
```

```python
# Save all BMI reports to a text file.
filename = "bmi_reports.txt"

with open(filename, "w") as file:
     for person in range(1, 4):
         print(f"---Person{person}---")
         name = input("Enter name: ")
         age = int(input("Enter age: "))
         weight = float(input("Enter weight in kg: "))
         height_cm = float(input("Enter height in cm: "))

         height_m = height_cm / 100
         bmi = round(weight / (height_m ** 2), 1)

         if bmi < 18.5:
             category = "Underweight"
         elif 18.5 <= bmi <= 24.9:
             category = "Normal weight"
         elif 25.0 <= bmi <= 29.9:
             category = "overweight"
         else:
             category = "Obese"
         # Write to file AND print to screen
         report = f"BMI REPORT FOR {name.upper()}\nAge: {age} | BMI: {bmi} | Category: {category}\n{'-'*40}\n"
         file.write(report)
         print(report)
         
print(f"\n All reports saved to {filename}")
print("Check your jupyter folder - the. txt file is there!")
```
